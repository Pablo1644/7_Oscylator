Oscylator harmoniczny

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
x = sp.Function('x')
w = sp.Symbol('omega')
t = sp.Symbol('t')
w0 = sp.Symbol('omega_0', real = True, positive = True)
x0 = sp.Symbol('x_0', real = True, positive = True)
v0=sp.Symbol('v_0', real = True, positive = True)
```

```python
x,w,t,w0,x0
```

Równanie bez tłumienia

```python
eq = x(t).diff(t, t)+w0 * x(t)
eq
```

```python
osc1 = sp.Eq(eq, 0)
osc1
```

Oscylator z tłumieniem

```python
beta=sp.Symbol('beta')
beta
```

```python
eq_t = x(t).diff(t, t)+2*beta*x(t).diff(t)+w0 * x(t)
eq_t
```

```python
osc_t = sp.Eq(eq_t,0)
osc_t
```

```python
solution_t=sp.dsolve(eq_t,x(t),ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
solution_t
```

```python
to_plot_t=solution_t.rhs
to_plot_t
```

```python
to_plot_t.subs({x0: 1, v0: -1, w0: 2,beta: 5})
```

```python
sp.plot(to_plot_t.subs({x0: 1, v0: -1, w0: 2,beta: 5}), (t, -5, 5))
```

```python
sp.plot(to_plot_t.subs({x0: 1, v0: -1, w0: 0.5,beta: 3}), (t, -5, 5))
```

```python
sp.plot(to_plot_t.subs({x0: 1, v0: -1, w0: 3,beta: 2}), (t, -5, 5))
```

Oscylator z wymuszeniem

```python
A = sp.Symbol('A')
F = A*sp.cos(w*t)
F
```

```python
osc_w = sp.Eq(eq_t,F)
osc_w
```

```python
solution_w=sp.dsolve(osc_w,x(t),ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
solution_w
```

```python
to_plot_w = solution_w.rhs
to_plot_w
```

```python
y_val_t=to_plot_w.subs({x0: 1, v0: -1, w0: 2,beta: 5,A:1,w: 1})
y_val_t
```

```python
sp.plot(to_plot_w.subs({x0: 1, v0: -1, w0: 2,beta: 5,A:10,w: 4}), (t, -5, 5))
```
